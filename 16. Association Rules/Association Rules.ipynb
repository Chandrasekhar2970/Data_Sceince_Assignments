```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel("Online retail.xlsx", header=None)
df.head()
```

```python
transactions = []

for i in range(0, 7501):
    transactions.append([str(df.values[i, j]) for j in range(20)])
len(transactions)
```

```python
transactions[0]
```

```python
transactions[7500]
```

```python
from apyori import apriori
rules = apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3,
                min_length = 2, max_length = 2)
results = list(rules)
len(results)
```

```python
results
```

```python
print('Support:', results[0][1]) # Support
print('Base item:', results[0][2][0][0]) # Base item
print('Add item1:', results[0][2][0][1]) # Add item
print('Confidence:', results[0][2][0][2]) # Confidence
print('Lift:', results[0][2][0][3]) # Lift
```

```python
a = []
b = []
c = []
d = []
e = []

for i in range(0, 9):
    c.append(results[i][1]) # Support
    a.append(results[i][2][0][0]) # Base item
    b.append(results[i][2][0][1]) # Add item
    d.append(results[i][2][0][2]) # Confidence
    e.append(results[i][2][0][3])

d1 = pd.DataFrame(a)
d2 = pd.DataFrame(b)
d3 = pd.DataFrame(c)
d4 = pd.DataFrame(d)
d5 = pd.DataFrame(e)

new_data = pd.concat([d1, d2, d3, d4, d5], axis = 1)
new_data.columns = ['Base item', 'Add item', 'Support', 'Confidence', 'Lift']
new_data
```

From the above analysis of the transactions, the confidence of 'Ground beef' for buying 'Tomato Sauce' is 0.377.

Which means the probability of buying 'Ground beef' is more likely when 'Tomato sauce' is purchased.

And also the lift ratio of 'Fromage blanc' and 'honey' is more compared to all other transactions.
