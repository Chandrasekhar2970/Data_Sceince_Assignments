```python
# Imputing the outliers with median value

import numpy as np # type: ignore

def replace_outliers_with_median(df):
  """
  Replaces outliers in each column of a DataFrame with the median of that column,
  excluding the outliers themselves.

  Args:
    df: A pandas DataFrame.

  Returns:
    A new pandas DataFrame with outliers replaced by median values.
  """
  new_df = df.copy()

  # Iterate over numeric columns
  for col in new_df.select_dtypes(include=[np.number]):
    # Define IQR outlier threshold function (adjust multiplier as needed)
    def is_outlier(series):
      q1 = series.quantile(0.25)
      q3 = series.quantile(0.75)
      iqr = q3 - q1
      lower_bound = q1 - (1.5 * iqr)
      upper_bound = q3 + (1.5 * iqr)
      return ~((series >= lower_bound) & (series <= upper_bound))

    # Find outliers (boolean Series)
    outliers = is_outlier(new_df[col])
    
    # Replace outliers with median (excluding outliers in median calculation)
    median_val = new_df.loc[~outliers, col].median()
    new_df.loc[outliers, col] = median_val

  return new_df

X = replace_outliers_with_median(df.select_dtypes(include='int64'))
print(X)
```

#### Isolation Forest Algoritm

```python
from sklearn.ensemble import IsolationForest
model = IsolationForest(contamination='auto')

# fitting the model
model.fit(df_num)
```

```python
# Predict the outlier score
df['outlier_score'] = model.decision_function(df_num)
df['outlier'] = model.predict(df_num)
```

```python
df['outlier'].value_counts()
```

The rows with value '-1' is the outlier within the numerical columns.

```python
df = df[df['outlier'] != -1]
df.drop(columns=['outlier_score', 'outlier'], inplace=True)
```

```python
df.shape
```

```python
df_num = df[['age', 'fnlwgt', 'capital_gain', 'capital_loss', 'hours_per_week']]

# Standardization of numerical columns
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
for i in df_num.columns:
    df_num[i] = ss.fit_transform(df_num[[i]])
df_num.head()
```

```python
# LabelEncoding
label_encoding = df[['workclass', 'education', 'education_num', 'marital_status', 'occupation', 'race', 'sex', 'relationship', 'native_country']]
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for i in label_encoding.columns:
    label_encoding[i] = le.fit_transform(label_encoding[[i]])
label_encoding.head()
```

#### Data Partition

```python
X = pd.concat([df_num, label_encoding], axis = 1)
X.head()
```

```python
X.shape
```

```python
Y = df[['income']]
Y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Fitting the model
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model = lr.fit(X_train, Y_train)
```

```python
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)

from sklearn.metrics import accuracy_score

ac1 = accuracy_score(Y_train, Y_pred_train)
ac2 = accuracy_score(Y_test, Y_pred_test)

print('Training accuracy:', round(ac1, 3))
print('Testing accuracy:', round(ac2, 3))
```
