```python
import pandas as pd
df = pd.read_csv('sales_data_with_discounts.csv')
df.head()
```

```python
df.info()
```

```python
df['SKU'].value_counts()
```

```python
df_cont = df[['Volume', 'Avg Price', 'Total Sales Value', 
                    'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']]
df_cat = df[['Day', 'BU', 'Brand', 'Model']]
df_remaining = df[['Date', 'SKU', 'City']]
```

```python
# Standardization --> The values arranged in between -3 to +3.
from sklearn.preprocessing import StandardScaler
SS = StandardScaler()

df_cont['Volume'] = SS.fit_transform(df_cont[['Volume']])
df_cont['Avg Price'] = SS.fit_transform(df_cont[['Avg Price']])
df_cont['Total Sales Value'] = SS.fit_transform(df_cont[['Total Sales Value']])
df_cont['Discount Rate (%)'] = SS.fit_transform(df_cont[['Discount Rate (%)']])
df_cont['Discount Amount'] = SS.fit_transform(df_cont[['Discount Amount']])
df_cont['Net Sales Value'] = SS.fit_transform(df_cont[['Net Sales Value']])
```

```python
df_cont.head()
```

```python
# Scaling --> It will convert any kind of numerical data to the values in between 0 and 1.
df_cont_scaling = df[['Volume', 'Avg Price', 'Total Sales Value', 
                    'Discount Rate (%)', 'Discount Amount', 'Net Sales Value']]

from sklearn.preprocessing import MinMaxScaler
MM = MinMaxScaler()

df_cont_scaling['Volume'] = MM.fit_transform(df_cont_scaling[['Volume']])
df_cont_scaling['Avg Price'] = MM.fit_transform(df_cont_scaling[['Avg Price']])
df_cont_scaling['Total Sales Value'] = MM.fit_transform(df_cont_scaling[['Total Sales Value']])
df_cont_scaling['Discount Rate (%)'] = MM.fit_transform(df_cont_scaling[['Discount Rate (%)']])
df_cont_scaling['Discount Amount'] = MM.fit_transform(df_cont_scaling[['Discount Amount']])
df_cont_scaling['Net Sales Value'] = MM.fit_transform(df_cont_scaling[['Net Sales Value']])
df_cont_scaling.head()
```

```python
# Label Encoding --> for Categorical variables
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()

df_cat['Day'] = LE.fit_transform(df_cat[['Day']])
df_cat['BU'] = LE.fit_transform(df_cat[['BU']])
df_cat['Brand'] = LE.fit_transform(df_cat[['Brand']])
df_cat['Model'] = LE.fit_transform(df_cat[['Model']])
df_cat.head()
```

```python
# One Hot Encoding 
df_cat = df[['Day', 'BU', 'Brand', 'Model']]
df_cat.head()

from sklearn.preprocessing import OneHotEncoder
OHE = OneHotEncoder()

dummy = OHE.fit_transform(df_cat).toarray()

df_cat_DF = pd.DataFrame(dummy)

df_cat_DF.head()
```

```python
df_final = pd.concat([df_remaining, df_cat, df_cont], axis = 1)
df_final
```

```python
# Here, Standardization is used to remove the units of the continuous variables and return the values in between -3 and +3.
# Scaling is used to convert the continuous variables values into the values in between 0 and 1.
```
