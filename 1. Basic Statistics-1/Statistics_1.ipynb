```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('sales_data_with_discounts.csv')
df.head()
```

```python
df.info()
```

```python
# Statistical measures of all the numerical columns
print('Mode of BU:', df['BU'].value_counts())
print('Mode of Brand:', df['Brand'].value_counts())
print('Mode of Model:', df['Model'].value_counts())
```

```python
df['Volume'].describe()
```

```python
df['Avg Price'].describe()
```

```python
print("Average of Total Sales Value:", df['Total Sales Value'].mean().round(3))
print('Mode of Total Sales Value:', df['Total Sales Value'].median())
print('Standard deviation of Total Sales Value:', df['Total Sales Value'].std())
```

```python
print("Average of Discount Rate:", df['Discount Rate (%)'].mean().round(3))
print('Mode of Discount Rate:', df['Discount Rate (%)'].median())
print('Standard deviation of Discount Rate:', df['Discount Rate (%)'].std())
```

```python
print("Average of Discount Amount:", df['Discount Amount'].mean().round(3))
print('Mode of Discount Amount:', df['Discount Amount'].median())
print('Standard deviation of Discount Amount:', df['Discount Amount'].std())
```

```python
print("Average of Net Sales Value:", df['Net Sales Value'].mean().round(3))
print('Mode of Net Sales Value:', df['Net Sales Value'].median())
print('Standard deviation of Net Sales Value:', df['Net Sales Value'].std())
```

```python
# Histogram of all numerical columns
df['Avg Price'].hist()
print(df['Avg Price'].skew())
# The Avg Price column is positively skewed histogram.
```

```python
df['Total Sales Value'].hist()
print(df['Total Sales Value'].skew())
# The Total Sales Value column is also having positively skewed.
```

```python
df['Discount Rate (%)'].hist()
print(df['Discount Rate (%)'].skew())
# Discount Rate (%) is negatively skewed histogram.
```

```python
df['Discount Amount'].hist()
print(df['Discount Amount'].skew())
# Discount amount is positively skewed histogram.
```

```python
df['Net Sales Value'].hist()
print(df['Net Sales Value'].skew())
# Net Sales Value column produces positively skewed histogram
```

```python
# Bar chart analysis of Categorical columns
df2 = df.groupby('BU')['Volume'].sum()
df2.plot(kind = 'bar')
```

Among the different models of BU, the volume of FMCG is the highest.

```python
df3 = df.groupby('Brand')['Volume'].sum()
df3.plot(kind = 'bar')
```

Among the volume of the different Brands, Jeera is having the large amount of volume among others.

```python
df4 = df.groupby('Model')['Volume'].sum()
df4.plot(kind = 'bar')
```

Among the Volume of different models, Balaji Oil is having the large amount of Volume.

```python
df2 = df.groupby('BU')['Net Sales Value'].sum()
df2.plot(kind = 'bar')
```

Among the different models of BU, Mobiles is having the highest amount of Net sales value.

```python
df2 = df.groupby('Brand')['Net Sales Value'].sum()
df2.plot(kind = 'bar')
```

Among the each different brand, Orange is having the highest amount of Net value sales.

```python
df2 = df.groupby('Model')['Net Sales Value'].max()
df2.plot(kind = 'bar')
```

Among the different models exists, S-20 is having the highest Net sales value.

```python
df['Brand'].value_counts().plot(kind = 'bar')
```

```python
df1 = df.groupby('Day')['Net Sales Value'].max()
df1.plot(kind = 'bar')
```

Among the days of week, Saturday is having the highest amount of Net sales value.
