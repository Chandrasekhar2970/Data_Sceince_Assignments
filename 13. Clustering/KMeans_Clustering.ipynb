```python
df_num = df.drop(['ID#', 'cc1_miles', 'cc2_miles', 'cc3_miles', 'Award?'], axis = 1)
df_num.head()
```

```python
# Standardization
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

df_num = ss.fit_transform(df_num)
df_num = pd.DataFrame(df_num)
df_num.columns = ['Balance', 'Qual_miles', 'Bonus_miles', 'Bonus_trans', 
                  'Flight_miles_12mo', 'Flight_trans_12', 'Days_since_enroll']
df_num.head()
```

```python
df_cat = df[['cc1_miles', 'cc2_miles', 'cc3_miles', 'Award?']]
df_cat.head()
```

```python
# Label Encoding
from sklearn.preprocessing import LabelEncoder # type: ignore
le = LabelEncoder()

for i in df_cat.columns: # type: ignore
    df_cat[i] = le.fit_transform(df_cat[i]) # type: ignore
df_cat.head()
```

```python
X = df_num
```

```python
X.head()
```

### K-Means Clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3, n_init = 20)

X['Kmeans'] = kmeans.fit_predict(df_num)
X['Kmeans'].value_counts()
```

```python
df_num.head()
```

```python
from sklearn.metrics import silhouette_score

scores = []

for i in range(2,10):
    cluster = KMeans(n_clusters=i,n_init=20)
    X["Kmeans"] = cluster.fit_predict(df_num.iloc[:, :-1])
    Y = X["Kmeans"]
    scores.append(silhouette_score(df_num.iloc[:, :-1],Y))
    
    
import matplotlib.pyplot as plt
plt.scatter(range(2,10),scores)
plt.plot(range(2,10),scores,color='red')
plt.xlabel("cluster numbers")   
plt.ylabel("silhoutte scores")
plt.show()
```

```python
X.head()
```

```python
# It is found that, the best cluster formation would be 3 clusters
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 4, n_init = 20)

X['Kmeans'] = kmeans.fit_predict(df_num.iloc[:, :-1])
X['Kmeans'].value_counts()
```

```python
# Using DBSCAN method,  have removed the outliers present in the dataset
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps = 5.45, min_samples = 3)

dbscan.fit(df_num.iloc[:, :-1])
dbscan.labels_
df_num['dbscan']  = dbscan.labels_
df_num['dbscan'].value_counts()
```

```python
# Combining the data to remove the rows with outliers
df_final = pd.concat([df_cat, df_num], axis = 1)
df_final.head()
```

```python
# Selecting the rows which contains other than '-1' in the 'dbscan' column of the dataset
df_final = df_final[df_final['dbscan'] != -1]
df_final.shape
```

```python
df_final.head()
```

```python
from sklearn.metrics import silhouette_score

scores = []

for i in range(2,10):
    cluster = KMeans(n_clusters=i,n_init=20)
    df_final["Kmeans"] = cluster.fit_predict(df_final.iloc[:, 4:11])
    Y = df_final["Kmeans"]
    scores.append(silhouette_score(df_final,Y))
    
    
import matplotlib.pyplot as plt
plt.scatter(range(2,10),scores)
plt.plot(range(2,10),scores,color='red')
plt.xlabel("cluster numbers")   
plt.ylabel("silhoutte scores")
plt.show()
```

```python
df_final.head()
```

#### Finalized model with 3 cluster formation

```python
from sklearn.cluster import KMeans # type: ignore
kmeans = KMeans(n_clusters = 3, n_init = 20)

df_final['Target'] = kmeans.fit_predict(df_final.iloc[:, 4:11])
df_final['Target'].value_counts()
```

```python
df_final.shape
```

```python
X_data = df_final.iloc[:, :11]
Y_data = df_final['Target']
```

```python
X_data.shape
```

```python
Y_data.shape
```

Data Partition

```python
from sklearn.model_selection import train_test_split # type: ignore
X_train, X_test, Y_train, Y_test = train_test_split(X_data, Y_data, test_size = 0.20, random_state = 10)
```

### Model fitting

1. KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier # type: ignore
model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, Y_train)
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)
```

```python
# metrics
from sklearn.metrics import accuracy_score # type: ignore
ac1 = accuracy_score(Y_train,Y_pred_train)
print("Training Accuracy score:" , round((ac1), 3))
ac2 = accuracy_score(Y_test,Y_pred_test)
print("Test Accuracy score:" , round((ac2), 3))
```

2. Support Vector Classifier

```python
from sklearn.svm import SVC # type: ignore
model = SVC(kernel = 'linear')

model.fit(X_train, Y_train)
Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)

# Metrics

from sklearn.metrics import accuracy_score # type: ignore
ac1 = accuracy_score(Y_train, Y_pred_train)
print('Training accuracy error:', round(ac1, 3))
ac2 = accuracy_score(Y_test, Y_pred_test)
print('Testing accuracy error:', round(ac2, 3))
```

3. Decision Tree Classifier

```python
# Model Fitting

from sklearn.tree import DecisionTreeClassifier # type: ignore
model = DecisionTreeClassifier(criterion='entropy',max_depth=11)
model.fit(X_train, Y_train)

Y_pred_train = model.predict(X_train)
Y_pred_test = model.predict(X_test)

# Metrics
from sklearn.metrics import accuracy_score # type: ignore
train_accuracy = accuracy_score(Y_train, Y_pred_train)
test_accuracy = accuracy_score(Y_test, Y_pred_test)

print("Training Accuracy:", round(train_accuracy, 3))
print("Testing Accuracy:", round(test_accuracy, 3))
```

Among the three classification models, Support Vector Classifier is having more training and test accuracies.

1. Training accuracy: 99.8%
2. Test accuracy: 99.5%
